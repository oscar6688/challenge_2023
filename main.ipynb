```python
# required libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy import ndimage
```

```python
# load in ChestMNIST data
data = np.load('chestmnist.npz')  # numpy.savez file type
```

```python
lst = data.files
print("list of datsets: ", lst)
```

```python
output_file_name = "Haosheng_Li_1.png"

# using train_images to load in an image
train_images = data['train_images']

img = train_images[0]
plt.title("Grayscale Image")
plt.imshow(img, cmap=plt.cm.gray);

plt.imsave(output_file_name, img, cmap=plt.cm.gray);
```

```python
output_file_name = "Haosheng_Li_2.png"

filtered_img = np.invert(img)
plt.title("Inverted Contrast Image")
plt.imshow(filtered_img, cmap=plt.cm.gray)

plt.imsave(output_file_name, filtered_img, cmap=plt.cm.gray);
```

```python
# using scipy.ndimage.gaussian_filter
output_file_name = "Haosheng_Li_3.png"

custom_filtered_img = ndimage.gaussian_filter(img, sigma=3)
plt.title("Custom Contrast Image")
plt.imshow(custom_filtered_img, cmap=plt.cm.gray)

plt.imsave(output_file_name, custom_filtered_img, cmap=plt.cm.gray);
```
